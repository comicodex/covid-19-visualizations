```
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
plt.style.use("fivethirtyeight")
from datetime import date, timedelta
```

```
df = pd.read_csv("https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv", parse_dates = ["Date"])

# Total confirmed cases
df["Total cases"] = df[["Confirmed", "Recovered", "Deaths"]].sum(axis=1)
df.head()
```

```
# Total worldwide cases grouped by date
worldwide_df = df.groupby(["Date"]).sum()
worldwide_df.head()
```

```
world_plot = worldwide_df.plot(figsize=(12, 5))
world_plot.set_xlabel("Date")
world_plot.set_ylabel("No. of cases worldwide")
world_plot.title.set_text("Worldwide covid-19")
plt.show()
```

```
# Comparing worldwide trend to Italy
it_df = df[df["Country"] == "Italy"].groupby(["Date"]).sum()

fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111) # 1x1 graph
ax.plot(worldwide_df[["Total cases"]], label="Worldwide")
ax.plot(it_df[["Total cases"]], label="Italy")
ax.set_xlabel("Date")
ax.set_ylabel("No. of total cases")
ax.title.set_text("Worldwide total cases vs Italy total cases")

plt.legend(loc="upper left")
plt.show()
```

```
# Italy daily cases and deaths (subtracting daily confirmed by previous day then shifting col down by 1)
it_df = it_df.reset_index()
it_df["Daily confirmed"] = it_df["Confirmed"].sub(it_df["Confirmed"].shift())
it_df["Daily deaths"] = it_df["Deaths"].sub(it_df["Deaths"].shift())

fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111) 

ax.bar(it_df["Date"], it_df["Daily confirmed"], color="b", label="Italy daily confirmed cases")
ax.bar(it_df["Date"], it_df["Daily deaths"], color="r", label="Italy daily deaths")
ax.set_xlabel("Date")
ax.set_ylabel("Italy daily confirmed cases")
ax.title.set_text("Italy daily cases and deaths")

plt.legend(loc="upper left")
plt.show();

```
